# 머닝러신의 분류
모델학습

1. 데이터를 모델에 넣고 결과를 낸다.
2. 결과를 정답과 비교해 다른 만큼 모델을 변경한다.
3. 특정 조건이 만족할 때까지 1,2를 반복한다.

손실함수

위 2번 단계에서 모델의 결과값이 실제 정답과 어떠 차이가 있는지 수치화할 필요가 있다. 즉, 이 수치화된 차이를 함수화한 것을 손실함수,비용함수 라고 한다. 다시말해, 손실함수는 모델의 선택에 대해 얼마나 손실이 일어났는지를 정의하는 함수로, 이 의미에 맞게 모델을 손실 함수의 값을 줄이는 방향으로 학습한다. 손실 함수는 보통 스칼라값으로 정의하는데, 모델이나 task에 따라 다양한 종류가 존재
Loss를 어떻게 설정하는냐가 모델 학습의 중요한 핵심 요인이 된다. 손실 함수중 가장 대표적인 함수로는 Mean Squared Error(MSE)를 들 수 있다.


# 머신러닝의 구분
지도학습

'지도한다' 라는 단어의 뜻을 생각해보면 이해하기 쉽다. 컴퓨터에게 '56세인 남성이 담배를 하루에 2갑씩 피는 사람은 폐암 걸릴 확률이 높고 25세의 비흡연자 남성은 폐암에 걸릴 확률이 높아!' 라고 지도한다고 생각하면 쉽게 이해할 수 있다. 지도 학습은 데이터가 주어졌을 때 무엇(x)으로 무엇(y)를 예축하고 싶을 때 사용한다. 쉽게 말해 공부를 시키는 데는 문제 지문과 답안이 둘 다 필요하다는 것이다.
어떠한 모델 y=f(x)에서 y가 실수형 값을 가질 때 풀어야 하는 문제를 '회귀 문제', 이때의 함수 f를 '회귀 모델'이라고 한다. 반면, 위 수식에서 y가 명목형 변수일 때, 다시 말해 특정 Class를 가질 때 풀어야 하는 문제를 '분류 문제'라고 하며 이때의 함수 f를 '분류 모델'이라 한다.


비지도학습

비지도학습은 지도학습의 반대개념, 데이터는 제공하지만 명확한 정답은 제공하지 않는 학습 방법

강화학습

최근에는 강화학습을 지도학습과 비지도학습 외에 다른 종류로 부른 이유는 강화학습의 지니고 있는 학습의 특성이 지도학습, 비지도학습과 다르기 때문이다. 강화학습은 수많은 시뮬레이션을 통해 컴퓨터가 현재 상태에서 어떤 행동을 취해야 먼 미래의 보상을 최대로 할 것인지를 학습하는 알고리즘(예를 들어 알파고)
강화학습을 하려면 현재상태, 행동, 보상, 다음 상태가 있어야한다. 즉, 시뮬레이션된 연속적인 데이터의 값이 존재해야 한다.
